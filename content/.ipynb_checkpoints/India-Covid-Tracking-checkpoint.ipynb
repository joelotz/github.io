```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
%matplotlib inline
```

```python
# Read data from file 
df = pd.read_csv("https://api.covid19india.org/csv/latest/case_time_series.csv")
# convert "Date" to proper datetime
df['Date'] = pd.to_datetime(df['Date'] + '2020')
#data.index = pd.to_datetime(data['Date'])

# Preview the first 5 lines of loaded data
df.head(5)
```

```python
# Look at the data 
print("Dataframe shape: ", df.shape)
df.isnull().sum()
```

```python
f, ax = plt.subplots(figsize=(20,10))
sns.set(style='whitegrid')
sns.scatterplot(x=df['Date'],y=df['Total Confirmed'],data=df,linewidth=2, ax=ax)
```

```python
df.loc[df['Date'] == '2020-03-22']['Total Confirmed']
```

```python
#import matplotlib.dates as mdates
from matplotlib.dates import MonthLocator, DateFormatter

fig, ax = plt.subplots(figsize=(20,10))
ax.plot('Date','Total Confirmed',data=df, linewidth=4)

# format the ticks
ax.xaxis.set_major_locator(MonthLocator())
ax.xaxis.set_major_formatter(DateFormatter('%b'))

# https://matplotlib.org/3.2.1/tutorials/text/annotations.html
# https://matplotlib.org/3.2.1/tutorials/text/text_props.html
notes = [['2020-03-22','March 22 -\nLockdown 1.0 Begins'], ['2020-03-25','March 25 -\nLockdown 2.0 Begins'],
         ['2020-04-15','April 15 -\nLockdown 3.0 Begins'], ['2020-05-04','May 04 -\nLockdown 4.0 Begins'],
         ['2020-06-01','June 01 -\nLockdown 5.0 Begins']]

for note in notes:
    print(note)
    ax.annotate(note[1], size=16,
        xy=(note[0], df.loc[df['Date'] == note[0]]['Total Confirmed']), xycoords='data',
        xytext=(-50, 200), textcoords='offset points',
        bbox=dict(boxstyle='round', facecolor='#f5f5f5', edgecolor='b'),
        arrowprops=dict(facecolor='black', arrowstyle='simple', connectionstyle='arc3'),
        horizontalalignment='left', verticalalignment='top')




```

```python



ax.annotate('March 25 -\nLockdown Begins', size=16,
            xy=('2020-03-25', df.loc[df['Date'] == '2020-03-25']['Total Confirmed']), xycoords='data',
            xytext=(-50, 200), textcoords='offset points',
            bbox=dict(boxstyle='round', facecolor='#f5f5f5', edgecolor='b'),
            arrowprops=dict(facecolor='black', arrowstyle='simple', connectionstyle='arc3'),
            horizontalalignment='left', verticalalignment='top')

ax.annotate('March 22 -\nJanta Curfew Begins', size=16,
            xy=('2020-03-22', df.loc[df['Date'] == '2020-03-22']['Total Confirmed']), xycoords='data',
            xytext=(-150, 100), textcoords='offset points',
            bbox=dict(boxstyle='round', facecolor='#f5f5f5', edgecolor='b'),
            arrowprops=dict(facecolor='black', arrowstyle='simple', connectionstyle='arc3'),
            horizontalalignment='left', verticalalignment='top')
```
